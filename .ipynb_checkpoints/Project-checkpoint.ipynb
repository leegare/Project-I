### Fibonacci

```python
# Iterative Method
def fib_it(n):
    f = [0,1]
    for a in range(n-2):
        f.append(f[-2]+f[-1])
    return f

# Recursive Method
def fib_rec(n, f=[]):
    if len(f)==0:
        return fib_rec(n, [0,1])
    if len(f)<n:
        f.append(f[-2]+f[-1])
        return fib_rec(n, f)
    else:
        return f
```

#### The following script will iterate with increasing values of 50 until 3000, calculate that number of digits of the fibonacci sequence, record it's execution time for the iterative function and the recursive function and graph it. 

```python
import time
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

#m = int(input('n:'))
m=3000
N = np.arange(0, m, 50)   # 0,10,20,..,m
n_r = []                # List of time executed in recursive mode
n_i = []                # List of time executed in iterative mode
for x in N:
    #print(x)
    start_time = time.time()
    fib_it(x)
    n_i.append(time.time()-start_time)
    start_time = time.time()
    fib_rec(x)
    n_r.append(time.time()-start_time)

plt.plot(N, n_i, label='Iterative')
plt.plot(N, n_r, label='Recursive')
plt.legend(loc='best')
plt.ylim(ymin=0, ymax=0.004)
plt.xlabel('$n$')
plt.ylabel('$seconds$')
plt.show()
```

### Binary Search

```python
def binary_search(sorted_list, goal):
    length = len(sorted_list)
    left = 0
    right = length - 1

    while left <= right:
        if goal < sorted_list[right] and goal > sorted_list[left]:
            piv = int((right+left)/2)            # Pivot
            if goal == sorted_list[piv]:
                return piv
            elif goal > sorted_list[piv]:
                left = piv
            else: 
                right = piv
        # Edge Cases 
        elif goal == sorted_list[right] or goal == sorted_list[left]:
            if goal == sorted_list[right]:
                return right
            else: 
                return left
        else:
            # Goal is not in the sorted list
            break
    return -1
```

```python
sorted_list = [0,7,7,8,25,25]
goal = 25
fx = binary_search(sorted_list,goal)
print(sorted_list[fx], fx)
```

### Linked List

```python
class node:
    def __init__(self,value=None):
        self.value = value
        self.next = None
        
class linked_list:
    def __init__(self):
        self.root = node()     # Head Node, Pointer to our first element on the list
        
    def append(self,value):
        new_node = node(value)
        current = self.root    # Start at the head prior to iteration
        while current.next is not None:
            current = current.next
        current.next = new_node
    
    def insert(self,value,slot):
        new_node = node(value)
        current = self.root
        slot_no = 0
        while current.next is not None:
            current = current.next
            if slot_no==slot:
                next_node = current.next
                current.next = new_node
                current = current.next
                current.next = next_node
            slot_no += 1
        
    def remove(self, d):
        this_node = self.root
        prev_node = None
        while this_node.next is not None:
            prev_node = this_node
            this_node = this_node.next
            if this_node.value == d:
                prev_node.next = this_node.next
            else: 
                pass

    def print_list(self):
        current = self.root
        lst = []
        while current.next: 
            current = current.next
            lst.append(current.value)
        return lst
```

```python
linklater = linked_list()
linklater.append(5)
linklater.append(1)
linklater.append(3)
linklater.append(19)
linklater.append(7)
print('Current Linked List:',linklater.print_list())
linklater.insert(5,2) # Insert x in slot y+1
print('Linked List After insertion of a 5 in the 2nd slot:\n',linklater.print_list())
linklater.remove(5)
print('Linked List After removal of any 5s ',linklater.print_list())
```

### Queues

```python
class node:
    def __init__(self,value=None):
        self.value = value
        self.next = None
```

```python
class queue: 
    def __init__(self):
        self.root = node()     # Head Node, Pointer to our first element on the queue
        self.size = 0
        
    def isEmpty(self):
        return self.size == 0
    
    def enqueue(self,item):
        new_node = node(item)
        current = self.root    # Start at the head prior to iteration
        while current.next is not None:
            current = current.next
        # Add item in the last slot (Not the head)
        current.next = new_node
        self.size += 1
        
    def dequeue(self):
        current = self.root
        self.root = current.next
        self.size -= 1
        
    def size(self):
        return self.size

    def print_queue(self):
        current = self.root
        lst = []
        while current.next: 
            current = current.next
            lst.append(current.value)
        return lst
```

```python
# Initialization
q1 = queue()
print('Empty?',q1.isEmpty())

# Queue Enqueue
q1.enqueue(1), print('\nQueue:',q1.print_queue()), 
q1.enqueue(2), print('Queue:',q1.print_queue()), 
q1.enqueue(3), print('Queue:',q1.print_queue()), 
q1.enqueue(4), print('Queue:',q1.print_queue()), 
q1.enqueue(6)

print('Queue:',q1.print_queue())
print('\nEmpty?',q1.isEmpty(), '! q1 has',q1.size, 'elements')

# Dequeue:
q1.dequeue(), print('\nQueue:',q1.print_queue())
q1.dequeue(), print('Queue:',q1.print_queue())
q1.dequeue(), print('Queue:',q1.print_queue())
q1.dequeue(), print('Queue:',q1.print_queue())
q1.dequeue()

print('\nEmpty?',q1.isEmpty())
```

### Stacks

```python
class stack: # LIFO
    def __init__(self):
        self.root = node()     # Head Node, Pointer to our first element on the queue
        self.size = 0
        
    def isEmpty(self):
        return self.size == 0
    
    def stack_in(self,item):
        new_node = node(item)
        current = self.root    # Start at the head prior to iteration
        while current.next is not None:
            current = current.next
        # Add item in the last slot (Not the head)
        current.next = new_node
        self.size += 1
        
    def stack_out(self):
        current = self.root
        for item in range(self.size-1):
            current = current.next
        current.next = None
        self.size -= 1
        
    def size(self):
        return self.size

    def print_stack(self):
        current = self.root
        lst = []
        while current.next: 
            current = current.next
            lst.append(current.value)
        return lst
```

```python
# Initialization
s1 = stack()
print('Empty?',s1.isEmpty())

# Stack In
s1.stack_in(1), print('\nStack:',s1.print_stack()), 
s1.stack_in(2), print('Stack:',s1.print_stack()), 
s1.stack_in(3), print('Stack:',s1.print_stack()), 
s1.stack_in(4), print('Stack:',s1.print_stack()), 
s1.stack_in(6)

print('Stack:',s1.print_stack())
print('\nEmpty?',s1.isEmpty(), '! s1 has',s1.size, 'elements')

# Stack Out
s1.stack_out(), print('\nStack:',s1.print_stack())
s1.stack_out(), print('Stack:',s1.print_stack())
s1.stack_out(), print('Stack:',s1.print_stack())
s1.stack_out(), print('Stack:',s1.print_stack())
s1.stack_out()
print('Stack:',s1.print_stack())
print('\nEmpty?',s1.isEmpty(), '! s1 has',s1.size, 'elements')
```

#### This method uses 2 functions, one to get n bernoulli events. Then the process will iterate the n bernoulli events m times. This is also the slowest method

```python
import pandas as pd
# Get n Bernoulli events
def bernoulli_trial(pb=0.5,nb=1000):
    return [pb<np.random.rand(1)[0] for tr in range(nb)]
# Returns the number of successes 
def flip_coins(tr):
    successes = [x for x in tr if x == True]
    return len(successes)


m=5000       # Number of sets
n=1000       # Number of trials
p=0.5        # Probability of coin toss
binomial_dist=[]
random.seed(0)

# Binomial trial for m times 
for t in range(m):
    # Get a set of n bernoulli trials and return with a list of True and False
    trial_set = bernoulli_trial(p,n)
    # Get the number of successes
    X = flip_coins(trial_set)
    binomial_dist.append(X)

    
# Plot histogramme
bd = pd.DataFrame(binomial_dist)
mini = bd.min()
maxi = bd.max()
plt.hist(binomial_dist, bins=np.arange(mini-1,maxi+1))
plt.xlabel('# of Successes')
plt.ylabel('# of Occurrences')
plt.show()
```

```python
import numpy as np
import pandas as pd

def bernoulli_trial(probability_of_yes):
    random_number = np.random.rand(1)[0]
    return probability_of_yes > random_number

def flip_a_fair_coin():
    return bernoulli_trial(0.5)


### This is an example of a binomial trial
flips = 1000
# 1000 True and Falses
coin_flip_binomial_trial = np.array([flip_a_fair_coin() for _ in range(flips)])
# Just for fun let's look at the precentage of heads
should_be_close_to_fifty_precent = (coin_flip_binomial_trial).sum() / flips
print(np.abs(should_be_close_to_fifty_precent - 0.5))

### Let's generalize a binomial trial
### a binomial trial is just a series of bernoulli_trials
def binomial_trial(n, p):
    return np.array([bernoulli_trial(p) for _ in range(n)])



### let's run a bunch of binomial trials to simulate the binomal distribution
bunch_of_binomial_trials = np.array([binomial_trial(1000, 0.5) for _ in range(5000)])
bunch_of_binomial_trials.shape # (5000, 1000)
# 5000 examples of flipping a coin 1000 times


binomial_dist = pd.DataFrame(bunch_of_binomial_trials.sum(1)) # Count the number of Yes (True) in each trial of 1000 flips
min = binomial_dist.min()
max = binomial_dist.max()
pd.DataFrame.hist(binomial_dist, bins=np.arange(min, max)) # This histogram has a similar shape to a binomial distribution
```

## Monty Hall Problem
#### The following script will approximate the probability of the 2 strategies commonly employed in the Monty Hall problem. 

```python
import random
import numpy as np

n_trials = 10000
n_wins = 0
n_switch_wins = 0
# Set the doors
doors = list(range(0, 3))

# Start set of trials: 
for trial in range(n_trials):
    
    
# Local Variables:
    # The list door_to_reveal will hold those door numbers that are likely to be revealed by the host
    door_to_reveal = []
    # The list doors_remaining will hold the door number that holds the alternative door for the player to swap.
    doors_remaining = []

    
    # The car gets hidden behind one of the 3 doors
    car = random.sample(doors,1).pop()

    
    # Participant selects a door randomly
    my_door = random.sample(doors,1).pop()

    
    # This for loop will define the two lists : Doors_to_reveal and Doors_remaining
    for d in doors:       
        if d != car:
            if d != my_door:
                door_to_reveal.append(d)
        elif car != my_door:
            doors_remaining.append(d)

            
    # Host selects a door that is not the door selected by the participant nor is a door that hides the car
    if len(door_to_reveal)>1:
        door_revealed = random.sample(door_to_reveal,1).pop()
        # Update doors_remaining
        for d in door_to_reveal:
            if d != door_revealed:
                doors_remaining.append(d)
    else: 
        door_revealed = door_to_reveal.pop()

        
    # Evaluate both strategies: switch doors and not switch them
    if car == my_door:
        n_wins += 1                      # Not switch

    if car == doors_remaining.pop():
        n_switch_wins += 1               # Switch

print('Wins:',n_wins/n_trials*100, '% vs Wins through switching doors', n_switch_wins/n_trials*100, '%')
```
